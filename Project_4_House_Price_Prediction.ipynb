Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn .datasets import fetch_california_housing
from sklearn import metrics
```

Importin the Boston House Price Dataset

```python
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
# print first 5 rows of our DataFrame
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.info()
```

```python
house_price_dataframe.isnull().sum()
```

```python
# getting the statictical measures of the data
house_price_dataframe.describe()
```

```python
# add the target(price) column to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.tail()
```

Understanding the correlation between various feature in teh dataset

1. Positive Correlation

2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the data
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and Target

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Spilting the data into Training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# traning the model X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction.shape)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print('R squared error : ', score_1)
print('Mean Absolute Error : ', score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predication Prices")
plt.title("actual Prices vs prediction Price")
plt.show()
```

Prediction on test data

```python
# accuracy for predition on test data
test_data_prediction = model.predict(X_test)
```

```python
# R cquared error
score_1 = metrics.r2_score(Y_test,test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('R squared error : ', score_1)
print('Mean Absolute Error : ', score_2)
```
